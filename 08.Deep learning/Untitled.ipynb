## Automatic Music Transcription(AMT)

- Predicting scores information from music audio
    - The primary score information is note (pitch, onset, duration)
    - Notes are arranged based on rhythm (beat), harmony (chord), and structure
    - Equivalent to automatic speech recogntion for speech signal
    
### Types of AMT Task

- Rhythm transcription (Time)
    - Onset detection
    - Tempo estimation
    - Beat tracking 

- Note transcription
    - Monophonic note
    - Polyphonic note
    - Expression detection (e.g. vibrato, pedal)
- Tonal Analysis
    - Key estimation
    - Chord recognition
    
- Timbre analysis
    - Multiple instrument recognition
    - Singing voice detection
    
- Structure analysis


### Score and Performance

### Audio-to-Score Alignment

- Temporal alignment between audio and score given a piece of music
    - Score
    - Audio (Waveform)
    - MIDI
- Algorithm Overview
    - Choose feature representation to compare
        - Midi : Convert audio to piano roll unsing a multi-pitch estimation algorithm
        - Audio feature : syntheszie audio from MIDI and extract feature
        
    - Feature Sequence1
    - Feature Sequence2
    - Similarity Matrix : Compute the local similarity
    - Dynamic Time Warping : Find the optimal path
    
- Audio feature representation
    - A commonly used choice for piano music is chroma
    - Similarity Matrix : Compute distance between frame-level feature

- Dynamic Time Warping
- Find the Minimum-Cost-Path
    - Dynamic Programming (DP)
$$
C_{k}(j) = O_{k}(j) + \min_{i}[C_{k}(i)+c_{ij}]
$$
