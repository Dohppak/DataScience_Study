## 서포트백터머신

- 마진을 최대화하면 분류 경계면을 찾아보자
- 분류 경계면을 $w^Tx+b$ 이라고 할때, 백터 W는 경계면과 수직인 법선의 백터이다.
- 이때 +와 -플레인 간의 관계는 ${ x }^{ + }={ x }^{ - }+\lambda w$ 이다
- $\lambda$는 $x^{+}$와 $x^{-}$사이의 관계로 다음과 같이 유도할 수 있다.
<br/><br/>

$$
% <![CDATA[
\begin{align*}
{ w }^{ T }{ x }^{ + }+b&=1\\ { w }^{ T }({ x }^{ - }+\lambda w)+b&=1\\ { w }^{ T }{ x }^{ - }+b+\lambda { w }^{ T }w&=1\\ -1+\lambda { w }^{ T }w&=1\\\\ \lambda =\frac { 2 }{ { w }^{ T }w }
\end{align*} %]]>
$$
