## 객체

프로그램이 어떤 작업을 수행하기 위해서는 __데이터__와 __데이터를 조작하는 행위__, 두 가지 요소가 필요하다. 일반적으로 데이터는 변수에 넣어서 사용하고 데이터를 조작하는 일은 함수를 통해한다

- Object(객체) : 서로 연관된 데이터와 그 데이터를 조작하기 위한 함수
- Memeber & Attribute (멤버 & 속성) : 집합의 원소가 되는 변수나 함수
- Method (메소드) : 객체의 속성인 함수

#### 사각형의 면적을 구하는 프로그램
1. 가로길이, 세로길이라는 두개의 데이터를 넣을 변수
2. 두 길이를 곱해서 면적을 구하는 함수


```python
h = 10
v = 20

def area(h,v):
    return h*v
a=area(h,v)
print(a)
```

```python
class Rectangle(object):
    def __init__(self,h,v):
        self.h=h
        self.v=v
        
    def area(self):
        return self.h*self.v
```

```python
r = Rectangle(10,20)
a = r.area()
print(a)
```

## 메소드 오버라이딩

메소드 오버라이딩은 여러 클래스에 걸쳐서 같은 이름의 매소드를 만드는 것이다. 예를들어서 모두 공통의 attack이라는 메소드가 있지만 다른 성격을 부여하고 싶을 수 있다. 이렇게 되면 부모 클래서에서 만든 매서드 정의를 자식 클래스에서 변경하여 설정한다.

```python
class Character(object):
    
    def __init__(self):
        self.life = 1000
        self.strength = 10
        self.intelligence = 10
        
    def attacked(self):
        self.life -= 10
        print(u"공격받음! 생명력 =", self.life) 
        
    def attack(self):
        print(u"공격!")  
```

```python
class Warrior(Character):
    
    def __init__(self):
        super(Warrior, self).__init__()
        self.strength = 15
        self.intelligence = 5
        
    def attacked(self):
        self.life -=30
        print(u"공격받음! 생명력", self.life)
    def attack(self):
        print(u"육탄 공격!") 
```

```python
class Wizard(Character):
    
    def __init__(self):
        super(Wizard, self).__init__()
        self.strength = 5
        self.intelligence = 15
        
    def attack(self):
        print(u"마법 공격!") 
```

```python
a = Character()
b = Warrior()
```

```python
b.attack()
```

```python
b.attacked()
```

```python
class overwatch:
    life = 100
    def setname(self, inputname): 
        #setname input name 
        self.name = inputname 
        # name input 
        print('이 캐릭터의 이름이 %s 로 정해졌습니다.' %(self.name))
    def attacked(self):
        self.life -= 10
        # life 10
        print('%s 이 공격을 받아 생명력이 %d 로 줄었습니다.' %(self.name, self.life))
```

```python
c = overwatch() 
c.setname('루시우') 
c.attacked()
```

1. C 라는 객체가 overwatch 클레스(붕어빵 틀)로 들어간다
2. C 라는 아이디를 가진 사람이 "루시우"라는 이름을 setname 함수에 입력으로 넣는다
    2-1. self.name = 루시우
3. c.name = name
4. c.name = 루시우

# 1. self

self 는 객체의 인스턴스 그 자체를 의미합니다. 대부분 객체지향 언어는 이걸 메소드에 안보이게 전달하지만, 파이썬의 경우에는 self 를 명시하고 메소드를 불러올떄 self를 자동으로 전달해야합니다.
<br/>

파이썬의 클래스는 기본적으로 외부에 독립되어 있는 함수에 껍데기를 씌운 것 입니다. self. 를 생각하게 되면 로컬과 인스턴스가 섞이게 됩니다. 

> 매서드의 첫번쨰 인자는 항상 self 여야한다. 왜냐 함수의 input이 하나도 없다면 self라도 전달해야하기 떄문이다

```python
class Foo:
    def func1():
        print("function1")
    def func2(self):
        print("function2")
```

```python
f = Foo()
f.func2()
```

```python
class Foo:
    def func1():
        print("function1")
    def func2(self):
        print(id(self))
        print("function2")
```

```python
f= Foo()
id(f)
```

```python
f2 =Foo()
id(f2)
```

```python
f2.func2()
```

```python
Foo.func1()
```

```python
class A(object):
    def __init__(self):
        self.x='Hello'
        
    def method_a(self, foo):
        print (self.x+''+foo)
```

```python
d1 = {'pi':3.14, 'e':2.718}
d2 = d1
d3 = d1.copy()
print(type(d1)==type(d2))
print(id(d1)==id(d2))
print(type(d1)==type(d3))
print(id(d1)==id(d3))
```

```python
class MyCounter:
    def __init__(self,value=0):
        self.counter = value
    def get(self):
        return self.counter
    def inc(self, value =1):
        self.counter += value
C1 = MyCounter()
print(C1.inc())
print(C1.inc(2))
print(C1.inc(3))
print(C1.get())
```

```python
def mystery(n):
    if n <= 0:
        return 0
    else:
        smaller = n-1
        return mystery(smaller) + n*n
```

```python
mystery(4)
```

```python
def mymin(L):
    if len(L) == 1:
        return L[1]
    elif L[0] < L[1]:
        return mymin(L[1:])
    else:
        return mymin(L[0:1]+L[1:])
```

```python
def mystery(self, Head):
    if Head is not None:
        some = Head
        while (some.next).next is not None:
            some = some.next
        some.next = None
    return Head
```
