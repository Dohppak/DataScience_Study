
## Pandas 인덱스 조작

### 데이터 프레임 인덱스 설정 및 제거

데이터 프레임에 인덱스로 들어가 있어야 할 데이터가 일반 데이터 열에 있거나, 열에 있어야 할 것이 인덱스로 되어 있을 수 있다.
- set_index : 기존 행 인덱스를 제거하고 데이터 열 중 하나를 인덱스로 설정
- reset_index : 기존 행 인덱스를 제거하고 인덱스를 마지막 데이터 열로 추가

```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(0)
df = pd.DataFrame(np.random.randint(1,10,(10,4)), columns = ["C1","C2","C3","C4"])
df
```

```python
df1 = df.set_index("C1")
df1
```

```python
df2 = df1.set_index("C2")
df2
```

```python
df1.reset_index()
```

```python
df1.reset_index(drop=True)
```

## 다중 인덱스

행이나 열에 여러 계층의 복수 인덱스를 설정할 수 있다.

```python
np.random.seed(0)
df3 = pd.DataFrame(np.random.randint(1,10,(10, 4)), columns =[["A","A","B","B"],["C1","C2","C3","C4"]])
df3
```

```python
df3.columns.names =["Cdx1","Cdx2"]
df3
```

```python
np.random.seed(0)
df4 = pd.DataFrame(np.random.randint(1,10,(8,4)),
                  columns=[["A","A","B","B"],["C","D","C","D"]],
                  index=[["M","M","M","M","F","F","F","F"],["ID" + str(i) for i in range(4)]*2]
                  )
df4.columns.names = ["Cdx1","Cdx2"]
df4.index.names = ["Rdx1","Rdx2"]
df4
```

## 행 인덱스와 열 인덱스의 교환

행 인덱스와 열 인덱스는 stack 이나 unstack 명령으로 교환가능하다.
- stack() : 열 인덱스를 행 인덱스로 벼환
- unstack() : 행 인덱스를 열 인덱스로 변환한다.
<br/>
<br/>


```python
df4.stack("Cdx1")
```

```python
df4.stack(0)
```

```python
df4.unstack("Rdx2")
```

```python
df3[("A","C1")]
```

```python
df4
```

```python
df4.loc[("M","ID0"),("A","C")]
```

## 인덱스 교환

인덱스의 순서를 바꾸고 싶으면 swaplevel 명령을 사용한다.
- swaplevel(i,j,axis)

i와 j는 교환하고자 하는 인덱스의 라벨(or 번호)이고 axis는 0일 때 행 인덱스, 1일때 열 인덱스를 뜻한다

```python
df4.swaplevel("Rdx1","Rdx2")
```

```python
df4.swaplevel("Cdx1","Cdx2",1)
```
